#### 5*. Написать на Python функцию для перемножения двух произвольных матриц, не используя NumPy.

```python
def matrix_multiply(x, y):
    res = []
    #проверка возможности перемножения (совпадения числа столбцов первой матрицы с числом строк второй)
    if len(x[0]) != len(y):
        print('Ошибка. Число столбцов первой матрицы должно равняться числу строк второй')
    else:
        for i in range(len(x)):
            res.append([])
            for j in range(len(y[0])):
                elem = 0
                for k in range(len(y)):
                    elem += x[i][k] * y[k][j]
                res[i].append(elem)
    return res
```

```python
#входные данные
A = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
B = [[1, 2], [3, 4], [5, 6]]
```

```python
#вывод результатов для наглядности
print(f'A:')
for row in A:
    print(row)
print(f'\nB:')
for row in B:
    print(row)    
print(f'\nA*B:')
for row in matrix_multiply(A, B):
    print(row)
print(f'\nB*A:')
for row in matrix_multiply(B, A):
    print(row)
```

```python
#проверка
print(f'A:\n{np.array(A)}\n')
print(f'B:\n{np.array(B)}\n')
print(f'A*B:\n{np.dot(np.array(A), np.array(B))}\n')
```

#### 8. Доказать, что матрица вырожденная

Вырожденная матрица - квадратная матрица, определитель которой равен 0.   
В данной матрице первая и вторая строка линейно зависимы. Согласно п.6 свойств определителя, если две строки (столбца) матрицы линейно зависимы, то определитель этой матрицы равен нулю. Следовательно, данная матрица - вырожденная.

```python
A = np.array([[-2, 7, -3], [4, -14, 6], [-3, 7, 13]])
print(f'A:\n{A}\n')
print(f'det:\n{round(np.linalg.det(A))}')
```

#### 9. Найти ранг матрицы:
